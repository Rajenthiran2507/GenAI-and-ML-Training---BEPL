```python
# Arithmetic operators
a = 10
b = 3
print(a + b)  # Addition: 13
print(a - b)  # Subtraction: 7
print(a * b)  # Multiplication: 30
print(a / b)  # Division: 3.3333333333333335
print(a // b)  # Floor Division: 3
print(a % b)  # Modulo: 1
print(a ** b)  # Exponentiation: 1000

# Assignment operators
x = 5
x += 3  # x = x + 3
print(x)  # Output: 8

# Comparison operators
print(a == b)  # Equal to: False
print(a != b)  # Not equal to: True
print(a > b)  # Greater than: True
print(a < b)  # Less than: False
print(a >= b)  # Greater than or equal to: True
print(a <= b)  # Less than or equal to: False

# Logical operators
p = True
q = False
print(p and q)  # Logical AND: False
print(p or q)  # Logical OR: True
print(not p)  # Logical NOT: False

# Bitwise operators
m = 5  # Binary: 0101
n = 3  # Binary: 0011
print(m & n)  # Bitwise AND: 1 (Decimal: 0001)
print(m | n)  # Bitwise OR: 7 (Decimal: 0111)
print(m ^ n)  # Bitwise XOR: 6 (Decimal: 0110)
print(~m)  # Bitwise NOT: -6 (Decimal: 1010)
print(m << 1)  # Bitwise Left Shift: 10 (Decimal: 1010)
print(n >> 1)  # Bitwise Right Shift: 1 (Decimal: 0001)
```

```python
# removing Dupliactes

def remove_duplicates(lst):
    result = []
    seen = set()

    for item in lst:
        if item not in seen:
            result.append(item)
            seen.add(item)

    return result
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Point):
            new_x = self.x + other.x
            new_y = self.y + other.y
            return Point(new_x, new_y)
        else:
            raise TypeError("Unsupported operand type for +")

    def __str__(self):
        return f"({self.x}, {self.y})"

point1 = Point(1, 2)
point2 = Point(3, 4)
point3 = point1 + point2

print(point3)
```

```python
#Multilevel Inheritance

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Animal speaks...")

class Mammal(Animal):
    def speak(self):
        print("Mammal speaks...")

class Dog(Mammal):
    def speak(self):
        print("Dog barks!")

# Creating instances
animal = Animal("Generic Animal")
mammal = Mammal("Generic Mammal")
dog = Dog("Tommy")

animal.speak()
mammal.speak()
dog.speak()
```

```python
# Accessing base class element in derived class

class BaseClass:
    def __init__(self):
        self.base_var = "Base Variable"

    def base_method(self):
        print("This is a method from the base class")


class DerivedClass(BaseClass):
    def __init__(self):
        super().__init__()  # Call base class's __init__ method
        self.derived_var = "Derived Variable"

    def derived_method(self):
        print("This is a method from the derived class")


# Creating an instance of the derived class
derived_obj = DerivedClass()

# Accessing base class variable and method using the derived class instance
print(derived_obj.base_var)
derived_obj.base_method()
```

```python
# Method Resolution Order

class A:
    def foo(self):
        print("A's foo")

class B(A):
    def foo(self):
        print("B's foo")
        super().foo()  # Calling parent class's method

class C(A):
    def foo(self):
        print("C's foo")
        super().foo()  # Calling parent class's method

class D(B, C):
    def foo(self):
        print("D's foo")
        super().foo()  # Calling parent class's method

# Creating instance
d = D()
d.foo()
```

```python
# Polymorphic Function

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2


# Function that accepts different shapes and computes their area
def calculate_area(shape):
    return shape.area()


# Create instances of different shapes
rectangle = Rectangle(5, 3)
circle = Circle(4)

# Call the calculate_area() function with different shapes
print(calculate_area(rectangle))
print(calculate_area(circle))
```

```python
# Overloading

# class Calculator:
#     def add(self, num1, num2):
#         return num1 + num2

#     def add(self, num1, num2, num3):
#         return num1 + num2 + num3


# calc = Calculator()

# print(calc.add(2, 3))
# print(calc.add(2, 3, 4))

# Working Around
class Calculator:
    def add(self, num1, num2, num3=None):
        if num3 is None:
            return num1 + num2
        else:
            return num1 + num2 + num3


calc = Calculator()

print(calc.add(2, 3))
print(calc.add(2, 3, 4))
```

```python
# Validate EMail Adress

import re

email = "example@example.com"

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

if re.match(pattern, email):
    print("Valid email address")
else:
    print("Invalid email address")
```

```python
# Split Text Into Sentsnces

import re

text = "This is sentence one. This is sentence two! And this is sentence three?"

pattern = r"(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s"

sentences = re.split(pattern, text)
print(sentences)  # Output: ['This is sentence one', 'This is sentence two', 'And this is sentence three']
```

```python
# EXtract Numebrs From A String

import re

text = "I have 3 apples and 2 bananas"

pattern = r"\d+"

numbers = re.findall(pattern, text)
print(numbers)  # Output: ['3', '2']
```

```python
# Remve Puncutatutions

import re

text = "Hello, World!"

pattern = r"[^\w\s]"

result = re.sub(pattern, "", text)
print(result)
```

```python
# Greedy Matching

import re

text = "Hello, World! How are you today?"

pattern = r"Hello.*you"

match = re.search(pattern, text)
if match:
    print(match.group())
```

```python
# Non Greedy Matching (Lazy Matching)

import re

text = "Hello, World! How are you today?"

pattern = r"Hello.*?you"

match = re.search(pattern, text)
if match:
    print(match.group())
```

```python
# re Flags

# re.IGNORECASE or re.I

import re

text = "Hello, World!"

pattern = r"hello"

match = re.search(pattern, text, re.IGNORECASE)
if match:
    print("Match found")
```

```python
# re.MULTILINE or re.M:

import re

text = "Hello\nWorld\n"

pattern = r"^World$"

match = re.search(pattern, text, re.MULTILINE)
if match:
    print("Match found")
```

```python
# re.DOTALL or re.S

import re

text = "Hello\nWorld"

pattern = r"Hello.World"

match = re.search(pattern, text, re.DOTALL)
if match:
    print("Match found")
```

```python
# re.VERBOSE or re.X

import re

text = "Hello World"

pattern = r"""
    Hello  # Match "Hello"
    \s     # Match whitespace
    World  # Match "World"
"""

match = re.search(pattern, text, re.VERBOSE)
if match:
    print("Match found")
```

```python
import re

sentence = "Check out our website at https://www.example.com for more information."

pattern = r"https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+"

url = re.findall(pattern, sentence)
print(url)

# The pattern used in this example is https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+. Let's break it down:

# https?://: Matches the protocol http:// or https://.
# (?:[-\w.]|(?:%[\da-fA-F]{2}))+: Matches one or more occurrences of either:
# [-\w.]: Matches any alphanumeric character, hyphen (-), or period (.) that may appear in the domain name.
# (?:%[\da-fA-F]{2}): Matches a percent sign (%) followed by two hexadecimal digits, which represents URL-encoded characters.
```

```python
# Creating A Menu File

import tkinter as tk

def new_file():
    print("Creating a new file...")

def open_file():
    print("Opening a file...")

def save_file():
    print("Saving the file...")

root = tk.Tk()

# Create a menu bar
menu_bar = tk.Menu(root)

# Create a file menu
file_menu = tk.Menu(menu_bar, tearoff=0)
file_menu.add_command(label="New", command=new_file)
file_menu.add_command(label="Open", command=open_file)
file_menu.add_command(label="Save", command=save_file)

# Add the file menu to the menu bar
menu_bar.add_cascade(label="File", menu=file_menu)

# Configure the root window to use the menu bar
root.config(menu=menu_bar)

root.mainloop()
```

```python
# GST Calculator

import tkinter as tk

def calculate_total():
    price = float(entry_price.get())
    gst_percentage = float(entry_gst.get())

    gst_amount = price * (gst_percentage / 100)
    total_amount = price + gst_amount

    label_total.config(text=f"Total Amount: {total_amount}", font=("Arial", 14))

root = tk.Tk()
root.geometry("300x250")
root.configure(padx=50, pady=50)

label_price = tk.Label(root, text="Price Amount:")
label_price.pack()

entry_price = tk.Entry(root)
entry_price.pack()

label_gst = tk.Label(root, text="GST Percentage:")
label_gst.pack()

entry_gst = tk.Entry(root)
entry_gst.pack()

button_calculate = tk.Button(root, text="Calculate", command=calculate_total)
button_calculate.pack()

label_total = tk.Label(root, text="Total Amount:", font=("Arial", 16))
label_total.pack()

root.mainloop()
```

```python
# Simple Calculator

import tkinter as tk

def button_click(value):
    current = entry_display.get()
    entry_display.delete(0, tk.END)
    entry_display.insert(tk.END, current + value)

def clear_display():
    entry_display.delete(0, tk.END)

def calculate_result():
    expression = entry_display.get()
    try:
        result = eval(expression)
        entry_display.delete(0, tk.END)
        entry_display.insert(tk.END, str(result))
    except:
        entry_display.delete(0, tk.END)
        entry_display.insert(tk.END, "Error")

root = tk.Tk()
root.title("Calculator")

entry_display = tk.Entry(root, font=("Arial", 20), width=15, justify=tk.RIGHT)
entry_display.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+'
]

row = 1
col = 0

for button in buttons:
    if button == '=':
        tk.Button(root, text=button, width=6, command=calculate_result).grid(row=row, column=col, padx=5, pady=5)
    elif button == '0':
        tk.Button(root, text=button, width=6, command=lambda: button_click(button)).grid(row=row, column=col, padx=5, pady=5, columnspan=2)
        col += 1
    else:
        tk.Button(root, text=button, width=6, command=lambda x=button: button_click(x)).grid(row=row, column=col, padx=5, pady=5)
    col += 1
    if col > 3:
        col = 0
        row += 1

clear_button = tk.Button(root, text="C", width=6, command=clear_display)
clear_button.grid(row=row, column=col, padx=5, pady=5)

root.mainloop()
```

```python
# Simple To Do

import tkinter as tk
from tkinter import messagebox

todos = []
completed_todos = []

def add_todo():
    todo = entry_todo.get()
    if todo.strip() != "":
        todos.append(todo)
        update_todo_list()
        entry_todo.delete(0, tk.END)
    else:
        messagebox.showwarning("Empty Todo", "Cannot add an empty todo.")

def delete_todo():
    selected_indices = listbox_todos.curselection()
    if len(selected_indices) == 0:
        messagebox.showwarning("No Selection", "No todo selected.")
    else:
        selected_index = selected_indices[0]
        todos.pop(selected_index)
        update_todo_list()

def complete_todo():
    selected_indices = listbox_todos.curselection()
    if len(selected_indices) == 0:
        messagebox.showwarning("No Selection", "No todo selected.")
    else:
        selected_index = selected_indices[0]
        todo = todos[selected_index]
        todos.pop(selected_index)
        completed_todos.append(todo)
        update_todo_list()
        update_completed_todo_list()

def update_todo_list():
    listbox_todos.delete(0, tk.END)
    for todo in todos:
        listbox_todos.insert(tk.END, todo)

def update_completed_todo_list():
    listbox_completed_todos.delete(0, tk.END)
    for todo in completed_todos:
        listbox_completed_todos.insert(tk.END, todo)

root = tk.Tk()
root.title("Todo App")
root.geometry("500x300")
root.resizable(False, False)

label_todo = tk.Label(root, text="Todo:")
label_todo.pack(pady=10)

entry_todo = tk.Entry(root)
entry_todo.pack()

button_add = tk.Button(root, text="Add Todo", command=add_todo)
button_add.pack(pady=10)

listbox_todos = tk.Listbox(root, selectmode=tk.SINGLE)
listbox_todos.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

listbox_completed_todos = tk.Listbox(root, selectmode=tk.SINGLE)
listbox_completed_todos.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

button_complete = tk.Button(root, text="Complete", command=complete_todo)
button_complete.pack(pady=5)

button_delete = tk.Button(root, text="Delete", command=delete_todo)
button_delete.pack(pady=5)

update_todo_list()
update_completed_todo_list()

root.mainloop()
```
